Part 1: Pre-processing the data

```python
#importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing the dataset
dataset = pd.read_csv('sonar.csv')
X = dataset.iloc[:, 0:60].values
Y = dataset.iloc[:, 60].values
```

```python
dataset.tail() #first row is taken for indexing
```

```python
X
```

```python
Y
```

```python
#encoding categorical values to numeric values where R is 1 and M is 0
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
Y = labelencoder.fit_transform(Y)
```

```python
#making the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
Y_test.size
```

Scaling is skipped as the 'Y' data contains only two categories.

Part 2: Creating the ANN

```python
#importing Keras 
import keras
from keras.models import Sequential
from keras.layers import Dense

#initialising the ANN
classifier = Sequential()
```

```python
#adding input layer, two hidden layers and output layer
classifier.add(Dense(activation="relu", input_dim=60, units=12, kernel_initializer="uniform"))
classifier.add(Dense(activation="relu", units=12, kernel_initializer="uniform"))
classifier.add(Dense(activation="relu", units=6, kernel_initializer="uniform"))
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))
```

```python
#compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#ANN with training set
classifier.fit(X_train, Y_train, batch_size = 10, epochs = 1000)
```

Part 3: Making predictions and Evaluating the model

```python
#test set results
Y_pred = classifier.predict(X_test)
Y_pred = (Y_pred > 8.5)

#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
```
